```python
data = pd.read_csv('ti8.csv')
data = data[['match_id','win','hero_id']]
```

```python
def get_hero_dict(group):
    """ get the hero name in one team
    Args:
        group: hero_id pandas.Series
    Returns:
        group: new added group hero_dict include heros 
    """
    match_dict = dict(group['hero_id'])
    group['hero_dict'] = str(list(match_dict.values()))
    return group
```

```python
data_hero = data.groupby(['match_id','win']).apply(get_hero_dict)
data_hero
```

```python
data_hero = data_hero[['match_id','win','hero_dict']]
```

```python
data_hero = data_hero.drop_duplicates()
```

```python
data_hero.head(6)
```

```python
def hero_count(group):
    """ get number of hero pick
        get number of two heros pick together
    Args:
        group: hero_dict pandas.Series
    Returns:
        hero_dict dict
        hero_group dict
    """
    hero_dict ={}
    hero_group = {}
    for each_team in group:
        for each_id in eval(each_team):
            if str(each_id) not in hero_dict:
                hero_dict[str(each_id)] = 1
            else:
                hero_dict[str(each_id)] += 1
                
            if str(each_id) not in hero_group:
                hero_group[str(each_id)] ={}
            for each_hero in eval(each_team):
                if each_hero ==  each_id:
                    continue
                elif str(each_hero) not in hero_group[str(each_id)]:
                    hero_group[str(each_id)][str(each_hero)] = 1
                else:
                    hero_group[str(each_id)][str(each_hero)] += 1
    return hero_dict,hero_group
```

```python
hero_dict,hero_group = hero_count(data_hero['hero_dict'])
```

```python
data_hero_win = data_hero[data_hero['win'] == True]
hero_dict_win,hero_group_win = hero_count(data_hero_win['hero_dict'])
```

```python
def make_WordCloud(chtext,save = False,name=''):
    """get wordcloud of hero name
    Args:
        chtext: heroname and it's frequence {'hero':num_pick}
        save  : save picture or not
        name  : title name
    Returns:
        None
    """
    from wordcloud import WordCloud  
    import matplotlib.pyplot as plt  
    font = 'C:\Windows\Fonts\STXIHEI.TTF'
    wc = WordCloud(
            background_color='white',
            font_path=font, 
            width=1000, height=860, 
            max_font_size=150,
            max_words = 50,)
    myfont = matplotlib.font_manager.FontProperties(fname=font)
    wc.generate_from_frequencies(chtext)   
    plt.title(name,fontproperties=myfont)
    plt.imshow(wc)  
    plt.axis("off")  
    plt.show() 
```

```python
def Normalization(input_list):
    """normalize the data from 0 to lens of data
    Args:
        inputlist: origin data
    Returns:
        normalized data
    """
    import numpy as np
    input_array = np.array(input_list)
    N_array = ((len(input_list))/(input_array.max()-input_array.min()))*\
    (input_array - input_array.min())
    
    return np.floor(N_array).tolist()#convert the float to int 
```

将原数据大小映射到目标区间

$out = (\frac{N_{max}-N_{min}}{O_{max}-O_{min}})*(O - O_{min})$

1. 节点的大小表示    **pick** 次数
2. 节点的颜色深浅表示 **获胜** 次数
3. 边的颜色表示     **组合** 次数

```python
import networkx as nx
import matplotlib.pyplot as plt
import matplotlib
plt.figure(figsize=(20,20))
G=nx.Graph()
a = []
num_fliter = 5 #只保留共同出场次数大于5次的英雄
for (hero_id1,hero_relationship) in hero_group.items():
    for (hero_id2, num_sameteam) in hero_relationship.items():
        if num_sameteam >= num_fliter:
            a.append([hero_json[hero_id1],hero_json[hero_id2],num_sameteam])
G = nx.Graph()
G.add_weighted_edges_from(a)
pos=nx.spring_layout(G)

matplotlib.rcParams['font.sans-serif'] = ['SimHei']   
matplotlib.rcParams['font.family']='sans-serif'

node_color = Normalization([hero_dict_win_cn[v] for v in G])
node_size = [0.4*(hero_dict_cn[v]**2) for v in G]

edge_color = Normalization([d['weight'] for (u , v , d) in G.edges(data=True)])

nx.draw_networkx_nodes(G, pos, node_color=node_color,node_size=node_size,cmap=plt.cm.Reds)
nx.draw_networkx_edges(G, pos, edge_color=edge_color ,edge_cmap=plt.cm.Blues,width = 2)
nx.draw_networkx_labels(G, pos, font_family='sans-serif',font_size = 12)

plt.axis('off')
plt.savefig("weighted_graph.png") # save as png
plt.show() # display
```
